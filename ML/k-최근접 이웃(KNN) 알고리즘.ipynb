# k-최근접 이웃(KNN)알고리즘

```python
import pandas as pd

data = pd.read_csv('../PJT/volleyball_fof.csv')
data
```

```python
data = data.drop(columns='Unnamed: 0')
```

```python
data.columns
```

```python
df = data.drop(['A팀_공격종합성공률', 'B팀_공격종합성공률'], 1)
```

```python
X = df.drop('target', axis = 1)
y = df['target']
```

## 1. 공격 종합만

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle = False, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import PolynomialFeatures
```

```python
k_param = range(3,15)
for k in k_param:
    pipeline = make_pipeline(
        KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    )
    pipeline.fit(X_train, y_train)
    print('k = ',k,'일 때 ',pipeline.score(X_train, y_train), pipeline.score(X_test, y_test))
```

```python
k_param = range(3,15)
for k in k_param:
    pipeline = make_pipeline(
        StandardScaler(),
        KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    )
    pipeline.fit(X_train, y_train)
    print('k = ',k,'일 때 ',pipeline.score(X_train, y_train), pipeline.score(X_test, y_test))
```

```python
# minmax보다 StandardScaler가 낫다
k_param = range(3,15)
for k in k_param:
    pipeline = make_pipeline(
        MinMaxScaler(),
        KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    )
    pipeline.fit(X_train, y_train)
    print('k = ',k,'일 때 ',pipeline.score(X_train, y_train), pipeline.score(X_test, y_test))
```

## 2. (오픈, 후위, 속공, 퀵오픈, 서브)

```python
df2 = data.drop(['A팀_오픈_성공률',
       'A팀_속공_성공률', 'A팀_퀵오픈_성공률', 'A팀_서브_성공률', 'A팀_블로킹_성공률', 'A팀_디그_성공률', 'B팀_오픈_성공률', 'B팀_속공_성공률', 'B팀_퀵오픈_성공률',
       'B팀_서브_성공률', 'B팀_블로킹_성공률', 'B팀_디그_성공률'], 1)

X2 = df2.drop('target', axis = 1)
y2 = df2['target']
```

```python
k_param = range(3,15)
for k in k_param:
    pipeline = make_pipeline(
        StandardScaler(),
        KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    )
    pipeline.fit(X_train2, y_train2)
    print('k = ',k,'일 때 ',pipeline.score(X_train2, y_train2), pipeline.score(X_test2, y_test2))
```

```python
k_param = range(3,15)
for k in k_param:
    pipeline = make_pipeline(
        MinMaxScaler(),
        KNeighborsClassifier(n_neighbors=k, n_jobs=-1)
    )
    pipeline.fit(X_train2, y_train2)
    print('k = ',k,'일 때 ',pipeline.score(X_train2, y_train2), pipeline.score(X_test2, y_test2))
```

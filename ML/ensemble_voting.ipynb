```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image
```

```python
df = pd.read_csv("../dataset/volleyball_fof.csv")
df.head(5)
```

```python
#공격종합만
df1 = df.drop(df.loc[:,'A팀_오픈_성공률':'A팀_디그_성공률'],1)
df1 = df1.drop(df1.loc[:,'B팀_오픈_성공률':'B팀_디그_성공률'],1)
df1.head(5)
```

```python
df2 = df.drop(['A팀_공격종합성공률', 'B팀_공격종합성공률'],1)
df2.head()
```

```python
##df1(공격종합만) - voting
```

```python
from sklearn.model_selection import train_test_split
```

```python
SEED = 0
X_train, X_test, y_train, y_test = train_test_split(df1.drop('target', axis=1), df1['target'], random_state=SEED)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import VotingClassifier, VotingRegressor
from sklearn.ensemble import RandomForestClassifier
```

```python
#no_params
```

```python
#lr 모델 생성
lr = LogisticRegression(random_state=0)
# lr 모델 학습
lr.fit(X_train, y_train)
#rdf 모델 생성
rdf_best = RandomForestClassifier(random_state = 0)
#rdf 모델 학습
rdf_best.fit(X_train, y_train)
#knn 모델 생성
knn = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)
#knn 모델 학습
knn.fit(X_train, y_train)
#voting classifier에 들어갈 파라미터 선언
single_models = [
    ('Random_forest', rdf_best), 
    ('knn', knn), 
    ('logistic_reg', lr)
]
#voting 모델 생성
voting_classifier = VotingClassifier(single_models, n_jobs=-1, voting='hard')
#voting 모델 학습
voting_classifier.fit(X_train, y_train)
#학습 데이터, 테스트 데이터 예측률 출력
voting_classifier.score(X_train,y_train), voting_classifier.score(X_test,y_test)
```

```python
#with_params
```

```python
lr = LogisticRegression(random_state=0)
lr.fit(X_train, y_train)
rdf_best = RandomForestClassifier(random_state = 0, max_depth = 8, max_features = 0.8,
            min_samples_leaf = 8, min_samples_split = 20, n_estimators = 1000)
rdf_best.fit(X_train, y_train)
knn = KNeighborsClassifier(n_neighbors=13, n_jobs=-1)
knn.fit(X_train, y_train)
single_models = [
    ('Random_forest', rdf_best), 
    ('knn', knn), 
    ('logistic_reg', lr)
]
voting_classifier = VotingClassifier(single_models, n_jobs=-1, voting='hard')
voting_classifier.fit(X_train, y_train)
voting_classifier.score(X_train,y_train), voting_classifier.score(X_test,y_test)
```

```python
SEED = 0
X_train, X_test, y_train, y_test = train_test_split(df2.drop('target', axis=1), df2['target'], random_state=SEED)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#no_params
```

```python
lr = LogisticRegression(random_state=0) #날 것
lr.fit(X_train, y_train)
rdf_best = RandomForestClassifier(random_state = 0) #날것
rdf_best.fit(X_train, y_train)
knn = KNeighborsClassifier(n_neighbors=5, n_jobs=-1) #날 것
knn.fit(X_train, y_train)
single_models = [
    ('Random_forest', rdf_best), 
    ('knn', knn), 
    ('logistic_reg', lr)
]
voting_classifier = VotingClassifier(single_models, n_jobs=-1, voting='hard')
voting_classifier.fit(X_train, y_train)
voting_classifier.score(X_train,y_train), voting_classifier.score(X_test,y_test)
```

```python
#with_params
```

```python
lr = LogisticRegression(random_state=0, max_iter = 1000)
lr.fit(X_train, y_train)
rdf_best = RandomForestClassifier(random_state = 0, max_depth = 8, max_features = 0.8,
            min_samples_leaf = 8, min_samples_split = 20, n_estimators = 1000)
rdf_best.fit(X_train, y_train)
knn = KNeighborsClassifier(n_neighbors=10, n_jobs=-1)
knn.fit(X_train, y_train)
single_models = [
    ('Random_forest', rdf_best), 
    ('knn', knn), 
    ('logistic_reg', lr)
]
voting_classifier = VotingClassifier(single_models, n_jobs=-1, voting='hard')
voting_classifier.fit(X_train, y_train)
voting_classifier.score(X_train,y_train), voting_classifier.score(X_test,y_test)
```

```python
# 3경기 이상 출전한 선수만 유지 (출전 세트 결측치가 3개 이상인 선수 제거)
subset = ['출전세트_1set', '출전세트_2set', '출전세트_3set', '출전세트_4set', '출전세트_5set']

# thresh = 3 : 정상 값(Not NaN)이 3개 이상만 유지
vol = vol.dropna(subset = subset, thresh = 3)
vol
```

#### 4️⃣ 공격 성공률
- 한개의 경기당, 팀별 공격 성공률
- 오픈, 후위, 속공, 퀵오픈, 서브 컬럼 각각 만들기
- 마지막에 전체 공격 컬럼만 살린 데이터 1개, 각각 컬럼 살린 데이터 1개로 총 2개의 데이터로 도출(다중 공산성 고려)

```python
vol.columns
```

```python
# 팀 전체 공격종합 성공률 평균
# sa = Successful attack
sa = vol.groupby(['팀명', '경기날짜'], as_index = False)[['공격종합_성공률']].mean().rename(columns = {'공격종합_성공률':'팀_공격종합성공률'})
sa
```

```python
# 팀 전체 오픈 성공률
# so = Successful Open
so = vol.groupby(['팀명', '경기날짜'], as_index = False)[['오픈_성공률']].mean().rename(columns = {'오픈_성공률':'팀_오픈_성공률'})
print(so.shape)

# 팀 전체 후위 성공률
sb = vol.groupby(['팀명', '경기날짜'], as_index = False)[['후위_성공률']].mean().rename(columns = {'오픈_성공률':'팀_후위_성공률'})
print(sb.shape)

# 팀 전체 속공 성공률
sf = vol.groupby(['팀명', '경기날짜'], as_index = False)[['속공_성공률']].mean().rename(columns = {'속공_성공률':'팀_속공_성공률'})
print(sf.shape)

# 팀 전체 퀵오픈 성공률
sqo = vol.groupby(['팀명', '경기날짜'], as_index = False)[['퀵오픈_성공률']].mean().rename(columns = {'퀵오픈_성공률':'팀_퀵오픈_성공률'})
print(sqo.shape)

# 팀 전체 서브 성공률
ss = vol.groupby(['팀명', '경기날짜'], as_index = False)[['서브_성공률']].mean().rename(columns = {'서브_성공률':'팀_서브_성공률'})
print(ss.shape)
```

#### 5️⃣ 블로킹 성공률
1. step 1) 블로킹 시도 합계 구하기
2. step 2) 블로킹 성공 합계 구하기
3. step 3) 성공을 시도로 나누어 성공률 구하기

```python
# 팀 전체 블로킹 성공률 구하기
# step 1) 블로킹 시도 합계 구하기
blocking_attempt= vol.groupby(['팀명', '경기날짜'], as_index = False)[['블로킹_시도']].sum().rename(columns = {'블로킹_시도':'팀_블로킹_시도_합계'})
blocking_attempt
```

```python
# 팀 전체 블로킹 성공률 구하기
# step 2) 블로킹 성공 합계 구하기
blocking_success= vol.groupby(['팀명', '경기날짜'], as_index = False)[['블로킹_성공']].sum().rename(columns = {'블로킹_성공':'팀_블로킹_성공_합계'})
blocking_success
```

```python
# 팀 전체 블로킹 성공률 구하기
# step 3) 성공을 시도로 나누어 성공률 구하기
blocking = blocking_attempt.merge(blocking_success, on = ['팀명', '경기날짜'])
blocking  # check

# 성공률 = 성공 / 시도
blocking['팀_블로킹_성공률'] = blocking['팀_블로킹_성공_합계'] / blocking['팀_블로킹_시도_합계'] * 100
blocking  # check

# 성공, 시도 컬럼 drop
blocking = blocking.drop(['팀_블로킹_시도_합계', '팀_블로킹_성공_합계'], axis = 1)
blocking  # check
```

#### 6️⃣ 디그 성공률, 디그 점유율
**[디그 성공률]**
1. step 1) 디그 시도 합계 구하기
2. step 2) 디그 성공 합계 구하기
3. step 3) 성공을 시도로 나누어 성공률 구하기

**[디그 점유율]**  
한개의 경기당 팀별로 평균 퍼센트

```python
# 팀 전체 디그 성공률 구하기
# step 1) 디그 시도 합계 구하기
digg_attempt= vol.groupby(['팀명', '경기날짜'], as_index = False)[['디그_시도']].sum().rename(columns = {'디그_시도':'팀_디그_시도_합계'})
digg_attempt
```

```python
# 팀 전체 디그 성공률 구하기
# step 2) 디그 성공 합계 구하기
digg_success= vol.groupby(['팀명', '경기날짜'], as_index = False)[['디그_성공']].sum().rename(columns = {'디그_성공':'팀_디그_성공_합계'})
digg_success
```

```python
# 팀 전체 디그 성공률 구하기
# step 3) 성공을 시도로 나누어 성공률 구하기
digg = digg_attempt.merge(digg_success, on = ['팀명', '경기날짜'])
digg  # check

# 성공률 = 성공 / 시도
digg['팀_디그_성공률'] = digg['팀_디그_성공_합계'] / digg['팀_디그_시도_합계'] * 100
digg  # check

# 성공, 시도 컬럼 drop
digg = digg.drop(['팀_디그_시도_합계', '팀_디그_성공_합계'], axis = 1)
digg  # check
```

```python
# 디그 점유율 평균
digg_share = vol.groupby(['팀명', '경기날짜'], as_index = False)[['디그_점유율']].mean().rename(columns = {'디그_점유율':'팀_디그_점유율'})
# print(digg_share.shape)
digg_share
```

#### 7️⃣ 리시브 성공률, 점유율
**[리시브 성공률]**
1. step 1) 리시브 시도 합계 구하기
2. step 2) 리시브 성공 합계 구하기
3. step 3) 성공을 시도로 나누어 성공률 구하기

**[리시브 점유율]**  
한개의 경기당 팀별로 평균 퍼센트  

```python
# 팀 전체 리시브 성공률 구하기
# step 1) 리시브 시도 합계 구하기
receive_attempt= vol.groupby(['팀명', '경기날짜'], as_index = False)[['리시브_시도']].sum().rename(columns = {'리시브_시도':'팀_리시브_시도_합계'})
receive_attempt
```

```python
# 팀 전체 리시브 성공률 구하기
# step 2) 리시브 성공 합계 구하기
receive_success= vol.groupby(['팀명', '경기날짜'], as_index = False)[['리시브_정확']].sum().rename(columns = {'리시브_정확':'팀_리시브_정확_합계'})
receive_success
```

```python
# 팀 전체 리시브 성공률 구하기
# step 3) 성공을 시도로 나누어 성공률 구하기
receive = receive_attempt.merge(receive_success, on = ['팀명', '경기날짜'])
receive  # check

# 성공률 = 성공 / 시도
receive['팀_리시브_성공률'] = receive['팀_리시브_정확_합계'] / receive['팀_리시브_시도_합계'] * 100
receive  # check

# 성공, 시도 컬럼 drop
receive = receive.drop(['팀_리시브_시도_합계', '팀_리시브_정확_합계'], axis = 1)
receive  # check
```

```python
# 리시브 점유율 평균
receive_share = vol.groupby(['팀명', '경기날짜'], as_index = False)[['리시브_점유율']].mean().rename(columns = {'리시브_점유율':'팀_리시브_점유율'})
# print(receive_share.shape)
receive_share
```

#### 4️⃣ ~ 7️⃣ 컬럼 원본에 merge하기

```python
vol.shape
```

```python
# df1 은 만든 1~3컬럼으로 저장, df2는 4~7컬럼으로 저장
# 4~7에 해당하는 여러 데이터 프레임 한번에 merge하기
from functools import reduce
import pandas as pd

dfs = [sa, so, sb, sf, sqo, ss, blocking, digg, digg_share, receive, receive_share]
print(len(dfs))
# Pre-processing
df_pp = reduce(lambda left, right: pd.merge(left, right, on = ['팀명', '경기날짜'], how = 'inner'), dfs)
```

```python
df_pp.shape
```

```python
df2 = vol.merge(df_pp, on = ['팀명', '경기날짜'])
df2
```
